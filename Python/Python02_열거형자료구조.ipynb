#Stack(FILO: Fisrst In Last Out)
- push: 데이터를 넣는 동작
- pop: 데이터를 끄집어 내는 동작
- List는 `stack`구조임

#Queue(FIFO: Fist In First Out)

-----------

```python
#List : 많이 사용하지만 무거운 객체
colors=['red','green','blue']
print(colors)
print(type(colors))
```

```python
#가장 뒤에 추가
colors.append('gold')
print(colors)
```

```python
#원하는 위치에 추가
colors.insert(1,'black')
print(colors)
```

```python
#여러 데이터를 추가
colors.extend(['white','gray'])
print(colors)
#연산자를 이용한 추가 방법은 혼란을 일으킬 수 있음
colors+= ['red']
print(colors)
```

```python
#[]를 제외하면 요소가 이상하게 저장됨
colors+='red'
print(colors)
```

```python
#해당 원소 index를 반환해줌
print(colors.index('red'))
print(colors.index('red',1))
print(colors.count('red')) #원소의 갯수
```

```python
#가장 뒤에 있는 요소를 끄집어냄
colors.pop()
```

```python
colors.pop()
```

```python
#원하는 위치의 요소를 끌어 낼 수 있음(stack, 위치 제거)
colors.pop(1)
print(colors)
```

```python
#해당 값을 제거(단순한 배열, 이름 제거)
colors.remove('gold')
print(colors)
```

```python
#값을 정렬(알파벳순으로)
colors.sort()
print(colors)
colors.reverse()
print(colors)
```

```python
color = {'apple':'red','banana':'yellow'}
print(color)
```

```python
#key에 해당되는 value
print(color['apple'])
#print(color[0])  #dict 타입은 index가 없음
#print(color['cherry'])  #없으면 error
```

```python
color['cherry']='red'
print(color)
```

```python
color['apple']='green'
print(color)
```

```python
#dictionary type안에 있는 값들을 list 형식으로 나열한 거임
#key만 뽑아냄
print(color.keys())
#value만 뽑아냄
print(color.values())

#list type안에 tuple type이 들어가있음
#item 단위로 뽑아냄
print(color.items())
```

```python
#for 문을 사용해서 데이터 가져오기
for k in color.keys():
  print(k)  
for tmp in color.values():
  print(tmp)
for i in color.items():
  print(i)
for (k,v) in color.items():
  print(k,v)
```

```python
#Dict에서 불필요한 아이템 삭제
del color['cherry']
print(color)
```

```python
#dict의 모든 아이템 삭제
color.clear()
print(color)
```

```python
#얕은 복사와 깊은 복사
a=[1,2,3]
#a의 주소를 copy (call by reference)
b=a
print(b)
```

```python
a[0]=38
print(a)
print(b)
```

```python
#call by value
a=[1,2,3]
b=a[:]
a[0]=38
print(a,b)
```

```python
#copy module을 사용하는 법(deep copy)
import copy

a=[1,2,3]
b=copy.deepcopy(a)
a[0]=38
print(a,b)
```
