```python
# Function

def Times(a,b):
  result=a*b
  return result
```

```python
print(Times)
```

```python
print(Times(10,20))
```

```python
#메모리 주소를 확인하는 function
print(id(Times))
```

```python
myTimes=Times #call by reference
print(myTimes(10,20))
```

```python
print(id(myTimes))
```

```python
#변수의 범위

g=10
x=20

def sum2(x,y):
  global g #전역변수 선언
  g+=1
  return x,y  #python에서는 tuple로 반환돼서 하나로 인식됨

print(sum2(x,20))
print(g)
```

```python
#Parameter

def Times(a=10,b=20): #default value 지정
  return a*b

print(Times(100,200))
print(Times(100)) #100*20
print(Times(b=200)) #10*200 
print(Times())
```

```python
def connectURL(protocol='https', server='naver.com',port='80'):
  address=protocol+'://'+server+':'+port
  return address


print(connectURL(port='80'))
```
