##OSI 7 layer

- `Physical` : 물리적인 연결 (ex 선, 잭, 이더넷 선은 왜 5인가? 같은 규약)
- `DataLink` : 전기적인 신호 (ex 5v로 신호를 주고 받겠다, 전기를 규정)
- `Network` : 신호들을 묶어서 정보 보내는 최소 단위(=패킷,  ex 데이터를 뭉처서 보낸다 101..)
  - `IP` : 빈병
- `Transport` : 신호 주고 받는 거를 control(ex 미국에 신호를 보냈는데 답장이 안 옴 ->기다림 -> 다시 보냄)  
  - `TCP`(Transmission Control Protocol) : 던지고 받는 방법
  - `UDP`(User Datagram Protocol) : 노빠꾸로 던진다
- `Session` :
