```python
plt.style.use('dark_background') #배경을 어둡게
rng=np.random.RandomState(0) #같은 seed 값을 지정하면 난수는 같은 값을 발생한다.
#RandomState(seed)는 random.seed(seed)와 달리 지역변수임으로 함수를 사용할때 다른점이 있다. (완벽히 이해하지 못했다 ㅠㅠ)


# rng.rand(5) #random값을 5개 생성
for maker in ['o',',',',','x','+','v','^','<','>','s','d']:
  plt.plot(rng.rand(5),rng.rand(5),maker,label='maker = {0}'.format(maker))
  plt.legend(numpoints=1) #legend에 올라간 점의 갯수
  plt.xlim(0,1.8) #x의 범위 
```

```python
plt.style.use('classic')

rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
color = rng.rand(100) #원의 색상
sizes = rng.rand(100) * 1000 #원의 크기
plt.scatter(x,y,s=sizes,c=color,alpha=0.3,cmap='cool') #alpha maker의 투명도
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T

plt.scatter(features[2],features[3],c=iris.target,s=features[3]*100,alpha=0.2,cmap='viridis') #target 품종확인
plt.xlabel(iris.feature_names[2])
plt.ylabel(iris.feature_names[3])
# iris.target_names[0]
```

```python
# 오차 시각화 하기
x=np.linspace(1,10)
dy=0.8 #error 값을 세로로 표시하기 위해서 
y=np.sin(x+np.random.randn(50))

plt.errorbar(x, y, yerr=dy,fmt='.k')
```
