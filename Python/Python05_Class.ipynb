```python
class Person:
  Name= '' #초기값 설정
  Age=0    #초기값 설정
  def __init__(self, name,age):
    self.Name=name 
    self.Age=age
  def __del__(self): #소멸자
    print('{0} 잘살다 갑니다'.format(self.Name))

  def PrintName(self): #class의 method들은 self를 반드시 붙여줌
    print(self.Name)
    print(self.Age)
```

```python
p=Person("황비홍",40)
p.PrintName()
del p
```

```python
#정적(Static) method
class CounterManager:
  InsCount=0

  def __init__(self):
    CounterManager.InsCount+=1
  def PrintInsanceCount(): #정적 메소드 변환
    print('Instance Count: ', CounterManager.InsCount)

  StaticCount= staticmethod(PrintInsanceCount)
```

```python
a = CounterManager()
b = CounterManager()
c = CounterManager()
```

```python
CounterManager.PrintInsanceCount()  #class 직접 사용하는 것은 금기사항
```

```python
p=CounterManager()
p.PrintInsanceCount()
```

```python
p = CounterManager()
p.StaticCount()
```

```python
# 상속의 심화
class Person:
  #초기값을 생성하지 않으면 자동으로 설정됨
  #초깃값을 선언하는 이유 : 가독성
  def __init__(self,name,phone):
    self.Name=name
    self.Phone=phone
  
  def PrintInfo(self):
    print('Info(Name: {0}, Phone {1}'.format(self.Name,self.Phone))

  def PrintPerson(self):
    print('Person(Name: {0}, Phone {1}'.format(self.Name,self.Phone))
```

```python
class Student(Person):
  def __init__(self,name,phone,subject,studentID):
    self.Name=name
    self.Phone=phone
    self.Subject=subject
    self.StudentID=studentID
```

```python
p=Person('박보검','010-1111-1111')
s=Student('민아','010-2222-2222','학생회장','0001')
```

```python
print(p.__dict__ )#해당 class 안에 내용을 dict type으로 보여줌
print(s.__dict__)
```

```python
#상속관계를 확인
issubclass(Student,Person)
```

```python
issubclass(Person,Student)
```

```python
class Student(Person):
  def __init__(self,name,phone,subject,studentID):
    self.Name=name
    self.Phone=phone
    self.Subject=subject
    self.StudentID=studentID

    Person.__init__(self,name,phone) #부모 class의 생성자 호출

  def PrintStudentData(self):
    print('Student(Name {0}, StudentID{1})'.format(self.Name,self.StudentID))
  
  def PrintPersonData(self):
    print('{0}님의 전공은 {1}입니다.'.format(self.Name,self.Subject))
```

```python
s=Student('민아','010-1111-1111','Computer','0001')
```

```python
s.PrintPersonData()
```

```python
#다중 상속

class Tiger:
  def Jump(self):
    print('호랑이 점프')
```

```python
class Lion:
  def Bite(self):
    print('사자의 물기')
```

```python
class Liger(Tiger,Lion):
  def Play(self):
    print('라이거의 놀기')
```

```python
l=Liger()
l.Jump()
l.Bite()
l.Play()
```
