```python
# 일차함수 (math 관련해서 이해 안 가면 EBS 드가쟈~~)

def linear_function(x):
  a=0.5
  b=2

  return (a*x)+b
```

```python
def quadratic_function(x):
  a=1
  b=-1
  c=-2

  return (a*x**2)+(b*x)+c
```

```python
print(quadratic_function(2))
```

```python
x=np.arange(-5,5,0.1)
y=quadratic_function(x)

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('quadratic function')
```

```python
def cubic_function(x):
  a = 4
  b = 0
  c = -1
  d = -8

  return a*x**3 + b*x**2 + c*x + d
```

```python
x = np.arange(-5,5,0.1)
y = cubic_function(x)

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Cublic Function')
```

# 논리 회로

```python
# AND
def AND(a,b):
  input=np.array([a,b])
  weights=np.array([0.4,0.4])
  bias=-0.6 
  value=np.sum(input*weights)+bias

  if value <= 0:
    return 0
  else:
    return 1
```

```python
print(AND(0,0))
print(AND(0,1))
print(AND(1,0))
print(AND(1,1))
```

```python
# AND gate
x1=np.arange(-2,2,0.01)
x2=np.arange(-2,2,0.01)
bias=-0.6

y=(-0.4 * x1- bias) /0.4
plt.axvline(x=0)
plt.axhline(y=0)

plt.plot(x1,y,'r--')
plt.scatter(0,0,color='orange',marker='o',s=150)
plt.scatter(0,1,color='orange',marker='o',s=150)
plt.scatter(1,0,color='orange',marker='o',s=150)
plt.scatter(1,1,color='black',marker='^',s=150)

#그래프 범위 제약
plt.xlim(-0.5,1.5)
plt.ylim(-0.5,1.5)
plt.grid()
```

```python
# OR
def OR(a,b):
  input=np.array([a,b])
  weights=np.array([0.4,0.4]) 
  bias=-0.3
  value=np.sum(input*weights)+bias

  if value <= 0:
    return 0
  else:
    return 1
```

```python
print(OR(0,0))
print(OR(0,1))
print(OR(1,0))
print(OR(1,1))
```

```python
# OR gate
x1=np.arange(-2,2,0.01)
x2=np.arange(-2,2,0.01)
bias=-0.3

y=(-0.4 * x1- bias) /0.4
plt.axvline(x=0)
plt.axhline(y=0)

plt.plot(x1,y,'r--')
plt.scatter(0,0,color='orange',marker='o',s=150)
plt.scatter(0,1,color='black',marker='^',s=150)
plt.scatter(1,0,color='black',marker='^',s=150)
plt.scatter(1,1,color='black',marker='^',s=150)

#그래프 범위 제약
plt.xlim(-0.5,1.5)
plt.ylim(-0.5,1.5)
plt.grid()
```

```python
# NAND
def NAND(a,b):
  input=np.array([a,b])
  weights=np.array([-0.5,-0.5]) 
  bias=0.7
  value=np.sum(input*weights)+bias

  if value <= 0:
    return 0
  else:
    return 1
```

```python
print(NAND(0,0))
print(NAND(0,1))
print(NAND(1,0))
print(NAND(1,1))
```

```python
# NAND gate
x1=np.arange(-2,2,0.01)
x2=np.arange(-2,2,0.01)
bias=-0.7

y=(-0.5 * x1- bias) /0.5
plt.axvline(x=0)
plt.axhline(y=0)

plt.plot(x1,y,'r--')
plt.scatter(0,0,color='black',marker='^',s=150)
plt.scatter(0,1,color='black',marker='^',s=150)
plt.scatter(1,0,color='black',marker='^',s=150)
plt.scatter(1,1,color='orange',marker='o',s=150)

#그래프 범위 제약
plt.xlim(-0.5,1.5)
plt.ylim(-0.5,1.5)
plt.grid()
```

```python
# XOR 
def XOR(x1,x2):
  s1=NAND(x1,x2)
  s2=OR(x1,x2)
  y=AND(s1,s2)

  return y
```

```python
print(XOR(0,0))
print(XOR(0,1))
print(XOR(1,0))
print(XOR(1,1))
```

# Activation Function

```python
# Step Function (계단함수)
def step_function(x):
  if x>0:
    return 1
  else:
    return 0
```

```python
print(step_function(0))
print(step_function(-1))
print(step_function(0.1))
```

```python
def step_function_for_numpy(x):
  y=x>0
  return y.astype(int)
```

```python
a=np.array([5,3,-4,2,0])
print(step_function_for_numpy(a))
```

```python
# Sigmoid Function

def sigmoid(x):
  return 1/(1+np.exp(-x))
```

```python
print(sigmoid(3))
print(sigmoid(10))
print(sigmoid(100))
print(sigmoid(-3))
print(sigmoid(-100))
```

```python
# Step function vs Sigmoid function
x=np.arange(-5,5,0.01)
y1=sigmoid(x)
y2=step_function_for_numpy(x)


plt.plot(x,y1,'r-',x,y2,'b--')
# plt.plot(x,y1,'r-')
# plt.plot(x,y2,'k-')
```

```python
# ReLU

def ReLU(x):
  if x>0:
    return x
  else:
    return 0
```

```python
print(ReLU(5))
print(ReLU(-5))
```

```python
x = np.arange(-5, 5, 0.01)
y = np.array([ReLU(i) for i in x])
plt.plot(x, y)
plt.show()
```

```python
# Softmax Function

def softmax(x):
  exp_x=np.exp(x)
  sum_exp_x = np.sum(exp_x)
  y=exp_x/sum_exp_x

  return y
```

```python
x=np.array([0.3,0.2,3.0,-1,2])
print(softmax(x))
print(np.sum(softmax(x))) #출력값의 총합은 1
```
