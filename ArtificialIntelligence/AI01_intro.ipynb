## Machine Learning
- 인간이 개발한 알고리즘을 컴퓨터 언어를 통해 기계에게 학습시키는 행위
- 가장 기본은 data, data-->computer(수학을 통해서 데이터를 수치 분석, 분류 같은 것을 함 = model) 
- 수학적인 방법(=Algorithm), 즉 데이터를 집어넣으면 algorithm을 통해 model 생성

# Machine Learning Algorithm

1. Supervised Learning (=지도학습)
* 문제와 정답 제공
  1. Classification(=분류)
    - Label과 Feature 을 집어넣으면 model이 생성
    - data를 학습하고 특징들을 정리한 모델에 테스트data를 넣어 성능확인
    - ex) 고양이 & 개 , 합격 & 불합격
  2. Regression(=회귀)
    - 값을 예측하는 것
    - ex) 키를 통해 발사이즈 예측하기

2. Unsupervised Learning(=비지도학습)
* 문제만 제공, 대표적인 예로는 Anomaly(변칙처리) : 패턴이 다 달라서 지도학습으로는 학습하기 힘듦
* 지도학습보다 정확도가 떨어짐
