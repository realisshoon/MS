# 데이터 전처리 
1. Scaling

```
  - 변수의 크기가 너무 작거나 큰 경우 Target에 미치는 영향이 제대로 표현되자 않을 수 있다.
  - sklearn을 사용함

```
  - Min-Max Scaling : 값의 범위를 0~1로 변경
  - Standard Scaling : z-score라고 하는 데이터를 통계적으로 표준정규분포화 시켜 스케일링한다.

2. Sampling

```
  - 클래스 불균형 문제는 분류를 목적으하는 데이터 셋에 클래스 라벨의 비율이 균형을 맞추지 않고 한쪽으로 치우친 경우를 말함

```

  - Over Sampling : 낮은 값을 올림
  - Under Sampling : 높은 값을 내림
  - Random Over, Under Sampling : 랜덤으로 값을 올리고 낮춤
  
3. Dimensionality Reduction (차원의 축소)
4. Categorical Variable to Numeric Variable

````python
abalone_path = join('./drive/MyDrive/MS/src','abalone.txt') #전복 데이터
column_path = join('./drive/MyDrive/MS/src','abalone_attributes.txt')  #속성을 저장한 .txt

abalone_columns = list() #column name을 만든다.
for l in open(column_path): #각 속성을 읽으면서 l에 저장함
  abalone_columns.append(l.strip()) #append() 뒤에 하나씩 추가함, strip() file을 가져올때 문자 뒤에 불필요한 요소를 제거, 공백도 제거함
#print(abalone_columns)

# Pandas를 이용해서 abalone 데이터를 읽어온다. 
data = pd.read_csv(abalone_path,header=None,names=abalone_columns) #1열을 header로 인식하기 때문에 None
data
````

````python
# head(), shape 를 사용해서 데이터를 살펴 본다. 
data.head()
````

````python
label = data['Sex'] 
del data['Sex'] #특정 column을 삭제
data.head()
````

````python
# Scaling (Min-Max)
data2 = (data - np.min(data)) / (np.max(data)-np.min(data))  #권장하지 않는 방식
data2
````

````python
# 1) 데이터 전처리 모듈의 불러오기 및 정의하기
from sklearn.preprocessing import MinMaxScaler

mMscaler = MinMaxScaler() 
mMscaler.fit(data)                #data를 minmaxscaler에 연결
data3 = mMscaler.transform(data)  #data를 변환하는 과정
data3
````

````python
#과정 간편화
mMscaler = MinMaxScaler()             
data4 = mMscaler.fit_transform(data) #fit_transform() 연결과 변환을 한 번에 하는 방법
````

````python
mMscaled_data = pd.DataFrame(data4, columns=data.columns) 
mMscaled_data.head()
````

````python
# 2. Standard Scaling
from sklearn.preprocessing import StandardScaler

sdscaler = StandardScaler()
sdscaled_data = sdscaler.fit_transform(data)

sdscaled_data = pd.DataFrame(sdscaled_data, columns=data.columns)
sdscaled_data.head()
````

````python
# Sampling
# 1. Random Over, Under Sampling

from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler

ros = RandomOverSampler()
rus = RandomUnderSampler()
````

````python
# Over Sampling
oversampled_data, oversampled_label = ros.fit_resample(data, label) 
oversampled_data = pd.DataFrame(oversampled_data, columns=data.columns)

# Under Sampling
undersampled_data, undersampled_label = rus.fit_resample(data, label)
undersampled_data = pd.DataFrame(undersampled_data, columns=data.columns)
````

````python
print('원본 데이터의 클래스 비율 \n{0}'.format(pd.get_dummies(label).sum()))
print('\nRandom Over 샘플링 결과 \n{0}'.format(pd.get_dummies(oversampled_label).sum()))
print('\nRandom Under 샘플링 결과 \n{0}'.format(pd.get_dummies(undersampled_label).sum()))
````

````python
# SMOTE

from imblearn.over_sampling import SMOTE
smote = SMOTE(k_neighbors=5)
````

````python
from sklearn.datasets import make_classification
data, label = make_classification(n_samples=1000, 
                                  n_features=2, 
                                  n_informative=2,
                                  n_redundant=0,
                                  n_repeated=0,
                                  n_classes=3,
                                  n_clusters_per_class=1,
                                  weights=[0.05,0.15,0.8],
                                  class_sep=0.8,
                                  random_state=2019)
````

````python
fig = plt.Figure(figsize=(12,6))
plt.scatter(data[:,0],data[:,1],c=label,linewidth=1,edgecolor='black')
````

````python
smoted_data, smoted_label = smote.fit_resample(data, label)
````

````python
print('원본 데이터의 클래스 비율\n{0}'.format(pd.get_dummies(label).sum()))
print('Smoted 데이터의 클래스 비율\n{0}'.format(pd.get_dummies(smoted_label).sum()))
````

````python
fig = plt.Figure(figsize=(12,6))
plt.scatter(smoted_data[:,0], smoted_data[:,1],c=smoted_label, linewidth=1,edgecolor='black')
````
