```python
from azure.storage.file import FileService
file_service=FileService(account_name='usercontainer91',account_key='')
```

```python
file_service.create_share('myshare') # file share 기본적으로 5GB씩 배정됨
```

```python
file_service.create_directory('myshare','sampledir') # mkdir
```

```python
from azure.storage.file import ContentSettings
file_service.create_file_from_path(
    'myshare',
    'sampledir',
    'myfile.jpeg',
    'mmayo.jpeg'
)
```

```python
# directory 내용의 확인
generator =file_service.list_directories_and_files('myshare')
for file_or_dir in generator:
  print(file_or_dir.name)
```

```python
# 파일을 가져오는 코드
file_service.get_file_to_path('myshare',
                              None,
                              'myfile',
                              'mmayo_DOWNLOAD.jpeg'
                              )
```

```python
# snapshot :사용자가 할 수 있는 백업..?
metadata ={'foo':'bar'}
snapshot = file_service.snapshot_share('myshare',metadata=metadata)
```

```python
share = list(file_service.list_shares(include_snapshots=True))
```

```python
#원하는 snapshot으로 복구?
directories_and_file = list(file_service.list_directories_and_files('myshare',
                                                                    snapshot='2023-04-20T05:48:10.0000000Z'))
for file_or_dir in directories_and_file:
  print(file_or_dir.name)
```

```python
#snapshot 제거
file_service.delete_share('myshare',snapshot='2023-04-20T05:48:10.0000000Z')
```

```python
# delete sharefolder
file_service.delete_file('myshare',None,'myfile')
```
