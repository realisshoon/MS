```python
import os,uuid #os : dir 생성/변경/삭제 등.. os의 역할을 함
from azure.identity import DefaultAzureCredential
from azure.storage.blob import BlobServiceClient, BlobClient,ContainerClient
```

```python
#Blob Service에 접속
connect_str='' #컨테이너의 연결문자열 key를 입력
blob_service_client1 = BlobServiceClient.from_connection_string(connect_str) #blob서버와 연결
```

```python
container_name='test2container' #컨테이너 이름 변수 설정

container_client = blob_service_client1.create_container(container_name) #컨테이너 생성하는 변수 설정
```

```python
#Create a local directory to hold blob data
local_path = './data'
os.mkdir(local_path) #Make a folder 

#Create a file in the local data directory to upload and download
local_file_name='welcome'+'.txt' #file 하나의 경로
upload_file_path= os.path.join(local_path,local_file_name) #file 전체 경로

#Write text to the file
file=open(upload_file_path,mode='w') #'Write mode'file open 
file.write('Welcome to Python!')
file.close() #file을 open 후 사용하고 나면  close 해야함
```

`binary` vs `text`

```python
blob_client= blob_service_client1.get_blob_client(  #위에서 만든 변수에서 container_client를 가져오라고 명령
                                    container=container_name,
                                    blob=local_file_name)
with open(file=upload_file_path,mode='rb') as data: # mode = 'rb' read binary
  blob_client.upload_blob(data) 
```

```python
# blob 파일의 리스트를 확인
print('\nListing blobs')

blob_list=container_client.list_blobs()
for blob in blob_list:
  print('\t'+blob.name)
```
